```python
#W4A11
n = int(input("Nhập một số nguyên dương n (n < 10^6): "))
if n <= 0 or n >= 10**6:
    print("Vui lòng nhập n hợp lệ: 0 < n < 10^6")
else:
    cnt = 0
    for i in range(2, n + 1, 2): 
        if n % i == 0:
            cnt += 1
    print(f"Số lượng ước số chẵn của {n} là: {cnt}")
```

```python
#W4A7
n = int(input("Nhập số nguyên dương n (n ≥ 2): "))
if n < 2:
    print("Vui lòng nhập n ≥ 2.")
else:
    max_prime = 1  
    for i in range(2, n + 1):
        if n % i == 0:
            is_prime = True
            for j in range(2, int(i**0.5) + 1):
                if i % j == 0:
                    is_prime = False
                    break
            if is_prime:
                max_prime = i  
    print(f"Ước số nguyên tố lớn nhất của {n} là: {max_prime}")
```

```python
import math
def is_prime(n):
    if n < 2:
        return False
    if n % 2 == 0:
        return n == 2
    k = int(math.sqrt(n))
    for i in range(3, k + 1, 2):
        if n % i == 0:
            return False
    return True

def check_positive_integer():
    while True:
        s = input()
        try:
            x = int(s)
        except ValueError:
            print("Sai: vui lòng nhập một số nguyên. Thử lại.")
            continue
        if x > 0:
            return x
        else:
            print("Sai: phải là số nguyên dương (> 0). Thử lại.")
            continue
n = check_positive_integer()
if is_prime(n):
    print("là số nguyên tố")
else:
    print("không là số nguyên tố")
```
