```python
#W5A3
import math
def is_prime (n):
    if n < 2:
        return False
    if n % 2 == 0:
        return n == 2
    k = int(math.sqrt(n))
    for i in range (3, k+1, 2):
        if n % i == 0:
            return False
    return True
try:
    n = int(input("Nhập vào số n:"))
    print (n,"là số nguyên tố" if is_prime(n) else "Không phải số nguyên tố")
except ValueError:
    pass
```

```python
#W5A8
def binary(a:int) -> str:
    res = ''
    while a != 0:
        res = str(a % 2) + res
        a = a//2
    return res
    
a,b = map(int,input().split())
s1 = binary(a)
s2 = binary(b)
while len(s1) < len(s2):
    s1 = '0' + s1
while len(s2) < len(s1):
    s2 = '0' + s2

cnt = 0
for i in range(len(s1)):
    if s1[i] != s2[i]:
       cnt += 1
print(cnt)
```
