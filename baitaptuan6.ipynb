```python
#W6A1
#Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
#Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]
A = list(map(int,input().split()))
S = []
for i in A:
    if i not in S:
        S.append(i)
print(S)
```

```python
#W6A2
#Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự
A = list(map(int,input().split()))
S = []
sum = 0
for x in A:
    sum += x
    S.append(sum)
print(S)
```

```python
#W6A3
#Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí
#(Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
A = tuple(map(int,input().split()))
k = int(input())
k = k % len(A)
A_rotated = A[k:] + A[k:]
print(A_rotated)
```

```python
#W6A4
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key 
#-> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key 
s = input().strip().split()
d = {}
for pair in s:
    key,value = pair.split(':')
    if key not in d:
        d[key] = []
    d[key].append(int(value))
print(d)
```
