```python
#bai7
def demuoc (n):
    dem = 0
    for i in range (1 , int ( n ** 0.5) + 1) :
        if n % i == 0  and i * i != n:
            dem += 2
        elif i * i == n:
            dem += 1
    return dem
n = int ( input ())
print (demuoc(n))
```
